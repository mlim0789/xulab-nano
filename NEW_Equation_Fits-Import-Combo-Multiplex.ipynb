# $P_{eff}$ fitting to two-compartment plasma PK

Form of bi-exponential plasma PK equation:

$$ C_V = C_0(\beta e^{-K_1 t}+(1-\beta)e^{-K_2 t}) $$

Form of equation for tumor accumulation: 
$$ C_T = \frac{APC_0\alpha}{1-H} \biggl[\frac{\beta}{\gamma_1} e^{-K_1 t} + \frac{1-\beta}{\gamma_2}e^{-K_2 t}- \biggl(\frac{\beta}{\gamma_1}+\frac{1-\beta}{\gamma_2}\biggr)e^{\frac{-AP}{\alpha V_T}t}\biggr ] $$
$$ \gamma_1 = AP-K_1 V_T \alpha $$ 
$$ \gamma_2 = AP - K_2 V_T \alpha $$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as fitter
%matplotlib inline


#IMPORTING VALUES FROM WORKBOOK

from openpyxl import load_workbook

#Put relative filepath to excel file here
book = load_workbook('C:/Users/Marc/Google Drive/UCB-UCSF/Xu lab/Project files/170602 Math models/Permeability Peff/200117 Data from Articles - Ct and PK formatted.xlsx')
sheetNames = book.sheetnames

#Enter index of sheet you want analyzed
worksheet = book[sheetNames[0]]    #want the first sheet of this excel file
colA = worksheet['A']

#GET TITLES

titlerow = 1       #titles are on Row 1
titleArr = []
colNum = 1
for colNum in range(1,50):
    colLetter = chr(colNum+65)   #series title is in Excel column B onwards
    col = worksheet[colLetter];
    if col[titlerow-1].value==None:    #stop loop when reach a column without any values
        break
    titleArr.append(col[titlerow-1].value)
print(titleArr)


#GET EXP

exprow = 2       #exponents are on Row 2
expArr = []
colNum = 1
for colNum in range(1,50):
    colLetter = chr(colNum+65)   #series title is in Excel column B onwards
    col = worksheet[colLetter];
    if col[exprow-1].value==None:    #stop loop when reach a column without any values
        break
    expArr.append(col[exprow-1].value)
print(expArr)


#GET C0

C0row = 3       #C0 are on Row 3
C0Arr = []
colNum = 1
for colNum in range(1,50):
    colLetter = chr(colNum+65)   #series title is in Excel column B onwards
    col = worksheet[colLetter];
    if col[C0row-1].value==None:    #stop loop when reach a column without any values
        break
    C0Arr.append(col[C0row-1].value)
print(C0Arr)


#GET ALL TIME VALUES

timerow = 5      #data are in Row 5
TUMtimeArr  = []
for i in range(timerow-1, 20):    #get timepoints stored in colA[4] = Excel row 5, onwards
    if colA[i].value==None:     #stop loop when reach a row without any values
        break
    TUMtimeArr.append(colA[i].value)    #data now stored all in one row in xdata array 
print(TUMtimeArr)
    
    
#GET CONCENTRATION VALUES

datarow = 5      #data are in Row 5
TUMconcArr = [[0 for height in  range(len(TUMtimeArr))] for width in range(len(titleArr))]   
          #empty 2D array with width of titles (i.e. number of lists) and height of timepoints (number of items / list)
counter = 0
for colNum in range(1,len(titleArr)+1):   #cycle through data in each Excel column
    colLetter = chr(colNum+65)            #chr(65) is A, and data is stored in columns B onwards (hence colNum + 65)
    col = worksheet[colLetter];          #select column in worksheet
    i = 0                                 #index for rows inside PKconc column list
    for j in range(datarow-1,datarow+len(TUMtimeArr)):     #extract data from current worksheet column 'col'  
        if col[j].value == None:            #stop if no data value found in current row 'j'. Excel data MUST BE "Number" type
            break
        #fill 'y' array row[counter] column[i] with value from current Excel column & row[j]
        TUMconcArr[counter][i] = col[j].value;  #fill value from worksheet column'col' row'j' into TUMconc column'counter' row'i'
        i = i + 1                      #cycle to next TUMconc row'i' before loop moves to next worksheet row'j'   
    counter = counter + 1              #cycle to next TUMconc column'counter' before loop moves to next worksheet column
print(TUMconcArr)


#IGNOR HEMATOCRIT AND INTERSTITIAL AVAILABLE VOLUME

alphaArr = [1,1,1,1,1] 
HArr = [0,0,0,0,0]  
betaArr = [1,1,1,1,1] 

#DEFINE MODEL

def fullmodel (x,p):   #function to calculate nanoparticle ceoncentration in tumor. "p" is P_eff
    A = 1 #flux area set to 1
    V_T = 100 #tumor vol set to 100 um3
    gamma1 = (A*p*C0*alpha*beta)/((1-H)*(A*p - K1*V_T*alpha))
    gamma2 = (A*p*C0*alpha)*(1-beta)/((1-H)*(A*p - K2*V_T*alpha))
    Ct = gamma1*np.exp(-K1*x)+gamma2*np.exp(-K2*x)-(gamma1+gamma2)*np.exp(-A*p*x/(alpha*V_T))  
    return Ct


#SOLVING FIT

PeffArr = []
Peff_errArr = []
Peff_RSD_Arr = []
#colors = ['red','blue','green','magenta','cyan','black']
#styles = ['ro','bo','go','mo','co','ko']
colors = ['red','magenta','black','green','blue','cyan','black']
styles = ['ro','mo','ko','go','bo','co','ko']

for j in range(0,len(titleArr)):  #solve fit and plot results for each column
#    plt.figure()
    TUMtime = TUMtimeArr    
    TUMconc = TUMconcArr[j]
    alpha = alphaArr[j]
    beta = betaArr[j]
    H = HArr[j]
    C0 = C0Arr[j]
    K1 = expArr[j]
    
    #Use values from PK fits

    par,cov = fitter.curve_fit(fullmodel,TUMtime,TUMconc)  
    #"par" is the optimal value of the unknown parameters
    #"cov" is the estimated covariance of "par". Array diagonals are the variance.

    #print(par)
    #print(np.sqrt(cov[0]))
    
    PeffArr.append(par)
    Peff_errArr.append(np.sqrt(cov[0]))
        
    Peff = PeffArr[j]
    Peff_err = Peff_errArr[j]
    Peff_RSD_Arr.append(100*Peff_err/Peff)

    #PLOTTING RESULTS

    plt.rcParams.update({'font.size': 15})

    plt.figure(1)
    #plt.subplot(2,1,1)       #2 rows, 1 column, first one  
    TUMxfit = np.linspace(0,TUMtime[len(TUMtime)-1],500)

    plt.title('Concentration in Tumor')
    plt.plot(TUMtime,TUMconc,styles[j])
    plt.plot(TUMxfit,fullmodel(TUMxfit,Peff[0]),linestyle='--',color=colors[j],label=titleArr[j])
    #    plt.legend(fontsize=12)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))     #puts legend box outside plot
    plt.xlabel('Time (hr)')
    plt.ylabel('Concentration')
plt.show()

from tabulate import tabulate
rowIDs3 = ["alpha","beta","H","Peff","SD","%SD"]
print(tabulate([alphaArr,betaArr,HArr,PeffArr, Peff_errArr,Peff_RSD_Arr], headers=titleArr,showindex=rowIDs3))
```
